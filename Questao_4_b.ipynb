```python
# Questão 4 - Exercicio b
import random
import pandas as pd
import matplotlib.pyplot as plt
from pandas.io.json import json_normalize
```

```python
#Informações de telemetria do carro 3
df = pd.read_json('telemetry.json', lines=True)
car3 = json_normalize( df[df['Car'] ==3 ]['telemetry'] )
car3.head()
```

```python
#Informações de RPM do carro 3 e gráfico separado
rpm = car3['RPM']
rpm.plot(kind='line', y=['RPM'], grid=True, color='green')
```

```python
#Informações da marcha do carro 3 e gráfico separado
gear = car3['Gear']
gear.plot(kind='line', y=['Gear'], grid=True, color='purple')
```

```python
# Reference: https://matplotlib.org/3.1.1/gallery/subplots_axes_and_figures/two_scales.html#sphx-glr-gallery-subplots-axes-and-figures-two-scales-py
# Gráficos de RPM e Gear sobrepostos do carro 3

fig, ax1 = plt.subplots()
color = 'tab:green'

# Gráfico RPM (eixo 1)
ax1.set_xlabel('Time (s)')
ax1.set_ylabel('RPM', color=color)
ax1.plot(rpm, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes

# Gráfico Gear (eixo 2)
color = 'tab:purple'
ax2.set_ylabel('Gear', color=color)
ax2.plot(gear, color=color, linewidth=2)
ax2.tick_params(axis='y', labelcolor=color)

# Plot do gráfico
fig.tight_layout()
plt.show()
```
